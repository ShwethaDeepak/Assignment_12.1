```python
#Charts to plot:
# 1. Create a pie chart presenting the male/female proportion

import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv("C:/pandas/titanic_original.csv")
#print(titanic)
count = titanic.groupby('sex').size()
print(count)
print (list(count))
count.plot.pie(figsize=(5,5))
plt.show()
```
